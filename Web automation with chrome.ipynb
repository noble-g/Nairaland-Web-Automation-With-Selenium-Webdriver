```python
import selenium
from selenium import webdriver as wd
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.wait import WebDriverWait
#from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from webdriver_manager.chrome import ChromeDriverManager
import chromedriver_autoinstaller
import chromedriver_binary

import time
from time import sleep
```

```python
#link = wd.Chrome(executable_path = 'C:\\Users\\User\\Downloads\\chromedriver')
# Either Or
link =wd.Chrome(ChromeDriverManager().install())

url = "https://www.nairaland.com/"
link.get(url)
```

```python
# define wait
wait = WebDriverWait(link, 10)
#wait.until(EC.presence_of_element_located((By.XPATH,'//*[@id="up"]/tbody/tr/td/a[1]')))

## read login details from txt file
txt = open('NL.txt')
lines = txt.readlines() 
```

```python
#login
login= link.find_element("xpath", '//*[@id="up"]/tbody/tr/td/a[1]')
login.click()

#time.sleep(5)
wait.until(EC.presence_of_element_located((By.XPATH,'//*[@id="up"]/tbody/tr/td/a[1]')))

username = link.find_element("xpath", "/html/body/div/table[2]/tbody/tr[2]/td/form/input[1]")
#username.send_keys(lines[0])
username.send_keys('noble4')

time.sleep(10)

password  = link.find_element("xpath", "/html/body/div/table[2]/tbody/tr[2]/td/form/input[2]")
password.send_keys(lines[1])
password.submit()

time.sleep(10)

#Refresh link
link.refresh()
time.sleep(10)

#open thread
thread = link.find_element("xpath", "/html/body/div/table[3]/tbody/tr[2]/td/a[1]")
thread.click()

time.sleep(10)
# Wait for the thread to load
#wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/div/table[3]/tbody/tr[2]/td/a[24]")))

#comment
reply = link.find_element("xpath", f"/html/body/div/p[3]/a[last()-1]")
reply.click()
comment = link.find_element("xpath", '//*[@id="body"]')
comment.click()
comment.send_keys('very insightful, thank you for sharing.')
comment.submit()

#like
like= link.find_element("xpath", "/html/body/div/table[2]/tbody/tr[2]/td/p[1]/a[3]")
like.click()

time.sleep(10)

#share
share= link.find_element("xpath", '/html/body/div/table[2]/tbody/tr[2]/td/p[1]/a[4]')
share.click()

time.sleep(10)

#follow OP
op = link.find_element("xpath","/html/body/div/table[2]/tbody/tr[1]/td/a[5]")
op.click()
link.find_element("xpath","/html/body/div/table[3]/tbody/tr/td/a").click()

time.sleep(10)

#log out
logout = link.find_element("xpath",'/html/body/div/table[1]/tbody/tr/td/a[13]').click()
link.close()
```

The snippet above logs in the forum with my login detail, opens the first thread, comment about the post, like the thread, share the thread and follow the poster of the thread and finally logs out.

```python
# Initialize
link =wd.Chrome(ChromeDriverManager().install())

url = "https://www.nairaland.com/"
link.get(url)

#login
login= link.find_element("xpath", '//*[@id="up"]/tbody/tr/td/a[1]')
login.click()

time.sleep(5)
#wait.until(EC.presence_of_element_located((By.XPATH,'//*[@id="up"]/tbody/tr/td/a[1]')))

username = link.find_element("xpath", "/html/body/div/table[2]/tbody/tr[2]/td/form/input[1]")
#username.send_keys(lines[0])
username.send_keys('noble4')

time.sleep(10)

password  = link.find_element("xpath", "/html/body/div/table[2]/tbody/tr[2]/td/form/input[2]")
password.send_keys(lines[1])
password.submit()
```

```python
link.refresh()
threads=  link.find_elements("xpath", "/html/body/div/table[3]/tbody/tr[2]/td/a")
thread_no = 0

for thread in threads:
    #open thread
    thread.click()
    time.sleep(10)
    thread_no += 1
    #print(thread_no)
    link.back()
    time.sleep(15)
    print(f"\nDone with thread {thread_no}")
    
```

You must have noticed the loop code above opens every thread in the home page on Nairaland. I could further include more actionn commands in the loop, say like or share every thread but to avoid beign noticed by the mod as a bot, I decided against that.

```python
link.refresh()
thread_no = 0
threads=  link.find_elements("xpath", "/html/body/div/table[3]/tbody/tr[2]/td/a")

for i in range(1,len(threads)):
    thread = link.find_element("xpath", f"/html/body/div/table[3]/tbody/tr[2]/td/a[{i}]")
    thread.click()
    time.sleep(10)
    thread_no += 1
    print(thread_no)
    link.back()
    time.sleep(10)
    print(f"\nDone with thread {thread_no}")
        
```
